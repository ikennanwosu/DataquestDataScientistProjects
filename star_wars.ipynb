# Cleaning and Exploring Star Wars Survery Data

While waiting for Star Wars: The Force Awakens to come out, the team at FiveThirtyEight became interested in answering some questions about Star Wars fans. In particular, they wondered: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?

The team needed to collect data addressing this question. To do this, they surveyed Star Wars fans using the online tool SurveyMonkey. They received 835 total responses, which you download from [their GitHub repository.](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey)

For this project, you'll be cleaning and exploring the data set.

#### Read the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Display all dataframe rows and columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

star_wars = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")
```

For this data cleaning project, the RespondentID is supposed to be a unique ID for each respondent, but from above we can see that it has a blank row. We will need to remove the row with invalid RespondentID.

```python
star_wars = star_wars[pd.notnull(star_wars.RespondentID)]
star_wars.head()
```

### Convert Columns Entries to Boolean Values and Rename Columns

#### Cleaning the First Two Columns

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
star_wars.iloc[:,:3].head()
```

#### Cleaning Next Six Columns

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars.iloc[:,:9].head()
```

#### Rename Column Names

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })
```

```python
star_wars.iloc[:,:9].head()
```

#### Cleaning up the next 6 columns

```python
# Convert the data type to float
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# Rename column names
new_cols = ["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]
old_cols = star_wars.iloc[:,9:15].columns

for i in range(len(new_cols)):
    star_wars = star_wars.rename(columns = {old_cols[i] : new_cols[i]})
```

```python
star_wars.iloc[:,:15].head()
```

Having cleaned up the ranking columns, it is easier to find the highest-ranked movie more quickly.

#### Display Average Rankings for Each Movie

```python
# Display the mean value of each Ranking
mean_rankings = star_wars.iloc[:,9:15].mean().sort_values(ascending=False)
mean_rankings
```

```python
N             = 6                              # the number of ranking categories
ind           = np.arange(N)                   # the x axis locations for each Ranking bars
width         = 0.5                            # the width of the bars
mean_rankings = star_wars.iloc[:,9:15].mean()  # mean value of the Ranking


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom', fontsize=20)


fig, ax = plt.subplots(figsize=(15,6))
rects1 = ax.bar(ind, mean_rankings, width)

# add some text for labels, title and axes ticks
ax.set_ylabel('Ranking', fontsize=20)
ax.set_title('Average Ranking', fontsize=20)
ax.set_xticks(ind + width / 2)

# Labels of the x axis ticks
rankings_label = tuple(star_wars.iloc[:,9:15].mean().index)
ax.set_xticklabels(rankings_label, fontsize=20)

autolabel(rects1)
plt.ylim(top=5)
plt.xticks(rotation=45)
plt.show()
```

We have so far cleaned the data, renamed some columns and computed the mean of the ranking of the movies. The plot shows that earlier episodes are rated much higher than newer episodes.

#### Summarize and visualize the number of people who have been each movie

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
N             = 6                                                # the number of ranking categories
ind           = np.arange(N)                                     # the x locations for each Ranking bars
width         = 0.5                                              # the width of the bars
sum_seen_movie = star_wars[star_wars.columns[3:9]].sum().values  # mean value of the Ranking


fig, ax = plt.subplots(figsize=(15,6))
rects1 = ax.bar(ind, sum_seen_movie, width)

# add some text for labels, title and axes ticks
ax.set_ylabel('Number of Seen Movies', fontsize=20)
ax.set_title('Sum of Star Wars Movies Seen', fontsize=20)
ax.set_xticks(ind + width / 2)

# Labels of the x axis ticks
rankings_label = tuple(star_wars.iloc[:,3:9].sum().index)
ax.set_xticklabels(rankings_label, fontsize=20)

autolabel(rects1)
plt.ylim(top=1000)
plt.xticks(rotation=45)
plt.yticks(fontsize=20)
plt.show()
```

The observation is quite different to the ranking of the movies. It seems the newer movies have been seen by more viewers than the earlier one. This could explain why the earlier movies were rated higher as more people who would have viewed the newer movies might have preferred and considered the older movies more original, and therefore rated them higher.

### Education
